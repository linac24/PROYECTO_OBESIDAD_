##### Análisis variables categoricas

A través de los siguientes gráficos podemos observar que no hay comportamientos atípicos o categorias extrañas en las variables categoricas

```python
D_cate=pd.concat([X_1.iloc[:, [0,4,5,8,9,11,14,15]],y],axis=1)
D_cate
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado datos con 9 variables categóricas

# Crear subgráficos 3x3
fig, axs = plt.subplots(3, 3, figsize=(12, 12))

# Iterar sobre las variables categóricas y graficar cada una
for i, var in enumerate(D_cate.columns):
    sns.countplot(x=var, data=D_cate, ax=axs[i // 3, i % 3])  # Calcular las coordenadas del subplot
    axs[i // 3, i % 3].set_title(f'Distribución de {var}')  # Establecer título
    axs[i // 3, i % 3].tick_params(axis='x', labelrotation=90)  # Rotar las etiquetas del eje x

# Ajustar el espacio entre los subgráficos
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

#### Análisis variables continuas

```python
D_conti=X_1.iloc[:, [1,2,3,6,7,10,12,13]]
D_conti
```

A continuación se observa el comportamiento de las variables continuas y teniendo en cuenta el criterio de valores atipicos fuera de los rangos intercuartilicos se identifica que las variables de Edad, Altura, Peso y NCP presentan valores atipicos. Sin embargo, las tres primeras variables en mencion representan caracteristicas normales que puede tener una persona, en tanto a la cuarta variable se ve necesario realizar un ajuste de transformación del tipo de variable.

```python
# Crear subgráficos 2x4
fig, axs = plt.subplots(2, 4, figsize=(15, 8))

# Iterar sobre las variables continuas y graficar cada una
for i, var in enumerate(D_conti.columns):
    sns.boxplot(y=D_conti[var], ax=axs[i // 4, i % 4], color='skyblue')  # Boxplot
    axs[i // 4, i % 4].set_title(f'Boxplot de {var}')  # Establecer título

# Ajustar el espacio entre los subgráficos
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

Transformación del Tipo de variable

```python
D_conti.dtypes
```

```python
D_conti.Age= D_conti.Age.round().astype(int)
D_conti.Height= D_conti.Height.round(2)
D_conti.Height

D_conti.Weight=D_conti.Weight.round()
D_conti.Weight

D_conti.FCVC=D_conti.FCVC.round().astype(int)
D_conti.FCVC

D_conti.NCP=D_conti.NCP.round().astype(int)
D_conti.NCP

D_conti.FCVC=D_conti.FCVC.round().astype(int)
D_conti.FCVC

D_conti.TUE=D_conti.TUE.round().astype(int)
D_conti.TUE
```

Es extraña la forma de las respuestas para el item: FCVC	Entero		¿Sueles comer verduras en tus comidas?  tiene valores entre 1 y 3, cuando la respuesta deberia ser dicotomica o indicar que significan los números entre 1 y3.

### Variables a estandarizar

Teniendo en cuenta que las variables continuas tienden a tomar valores entre 0 y 3, se propone realizar una transformación de las  variables'Age' y 'Weigth', pasando a un rango de cero a 3 aprox.

```python
import numpy as np

D_conti.Age.Transf=(D_conti.Age-np.mean(D_conti.Age))/np.std(D_conti.Age)
D_conti.Age.Transf
```

```python
D_conti.Age_2= D_conti.Age*3/70
D_conti.Age_2
```

```python
D_conti.Weight_2= D_conti.Weight*3/180
D_conti.Weight_2
```
